## Data Loader

```python
def preprocess(s):
    def replace(s):
        s = s.encode('utf-8').decode('ascii', 'ignore') # remove emoji?
        s = re.sub(r'\n', ' ', s)
        s = re.sub(r'&amp;', '', s)
        s = re.sub(r'\s+\.', '.', s)
        s = re.sub('\s+', ' ', s)
        return s.strip()
    s = replace(s)
    new_text = []
    for t in s.split():
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return ' '.join(new_text)

def loaddata(task):
    def load(dataset, typ):
        with open(os.path.join(VOL, 'datasets/text', task, f'{dataset}_{typ}.txt'), 'r') as f:
            lines = []
            for line in f:
                lines.append(line.rstrip())
        return np.array(lines)
    Xtrain, Ttrain = np.array([preprocess(s) for s in load('train', 'text')]), \
        load('train', 'labels').reshape(-1,1).astype(int)
    Xval  , Tval   = np.array([preprocess(s) for s in load('val', 'text')]), \
        load('val', 'labels').reshape(-1,1).astype(int)
    Xtest , Ttest  = np.array([preprocess(s) for s in load('test', 'text')]), \
        load('test', 'labels').reshape(-1,1).astype(int)
    
    return Xtrain, Ttrain, Xval, Tval, Xtest, Ttest

def loadlabels(task):
    mapping = os.path.join(VOL, 'datasets/text', task, 'mapping.txt')
    
    with open(mapping, 'r') as f:
        labels = {}
        for line in f.read().splitlines():
            split = line.split('\t')
            labels[int(split[0])] = split[1]
    return labels

def loadmel(task, dataset):
    audiodir = f'datasets/audio/output/{task}/{dataset}'
    return [np.load(f) for f in glob.glob(os.path.join(VOL, audiodir, f'*.npy'))]
```

## Network Training

```python
def t2i_loader(Xtrain, embed_mode='glove'):
    # TODO: check tokenizer + vocab for words like "don't"
    tokenizer = get_tokenizer('basic_english') # 'spacy'
    
    if embed_mode == 'default':
        vocab = np.array(list(set(w for s in Xtrain for w in tokenizer(s))))
        vocab = np.insert(vocab, 0, '<pad>')
        vocab = np.insert(vocab, 1, '<unk>')
        
        embeddings = None
        
        wtoi = {w:i for i,w in enumerate(vocab)}
        wtoi = defaultdict(lambda: wtoi['<unk>'], wtoi)
        itow = {i:w for w,i in wtoi.items()}

    elif embed_mode == 'glove':
        # https://medium.com/mlearning-ai/load-pre-trained-glove-embeddings-in-torch-nn-embedding-layer-in-under-2-minutes-f5af8f57416a
        dim = 50
        with open(os.path.join(VOL, 'embeddings', 'twitter', f'glove.twitter.27B.{dim}d.txt'), 'rt') as fi:
            full_content = fi.read().strip().split('\n')

        vocab, embeddings = [],[]
        for i, content in enumerate(full_content):
            split = content.split()
            i_word = split[0]
            i_embeddings = [float(v) for v in split[1:]]
            if len(i_embeddings) == dim: # some aren't equal?
                vocab.append(i_word)
                embeddings.append(i_embeddings)

        vocab = np.array(vocab)
        embeddings = np.array(embeddings)

        # insert '<pad>' and '<unk>' tokens
        vocab = np.insert(vocab, 0, '<pad>')
        vocab = np.insert(vocab, 1, '<unk>')

        pad = np.zeros((1, embeddings.shape[1]))
        unk = np.mean(embeddings, axis=0, keepdims=True)

        embeddings = np.vstack((pad,unk,embeddings))

        wtoi = {w:i for i,w in enumerate(vocab)}
        wtoi = defaultdict(lambda: wtoi['<unk>'], wtoi)
        itow = {i:w for w,i in wtoi.items()}
    else:
        raise ValueError(f'Incorrect {embed_mode=}')

    stoi = lambda x: [[wtoi[w] for w in tokenizer(s)] for s in x]
    
    return stoi, wtoi, itow, embeddings, vocab


def trainer(text=True, audio=True, text_mode='lstm', embed_mode='glove'):
    Xtrain, Ttrain, Xval, Tval, Xtest, Ttest = loaddata(task)
    if text:
        stoi, wtoi, itow, embeddings, vocab = t2i_loader(Xtrain, embed_mode)
        Xtrain, Xval, Xtest = list(map(stoi, [Xtrain, Xval, Xtest]))
        text_args = {'padding_idx': wtoi['<pad>'], 'activation_f': 'tanh'}
        if text_mode == 'lstm':
            text_args['lstm_args'] = {'lstm_hidden_dim': 64, 'n_lstm_layers': 1}
        else:
            text_args['cnn_args'] = [{'n_units': 8, 'window': 1}, {'n_units': 8, 'window': 3},
                                     {'n_units': 8, 'window': 5}, {'n_units': 8, 'window': 5}]
        text_args['embed_args'] =  {'embeddings': embeddings} if embed_mode == 'glove' \
            else {'num_embeddings': len(vocab), 'embedding_dim': 100}
    else:
        text_args = None
        
    if audio:
        Atrain = loadmel(task, 'train')
        Aval   = loadmel(task, 'val')
        Atest  = loadmel(task, 'test')
        max_mel_len = min(len(max(Atrain, key=len)), 1500)
        
        audio_args = {'n_inputs': (1, max_mel_len, Atrain[0].shape[-1]),
                      'conv_layers': [{'n_units': 8, 'shape': 3},
                                      {'n_units': 16, 'shape': 3},
                                      {'n_units': 32, 'shape': 3},
                                      {'n_units': 64, 'shape': 3}],
                      'activation_f': 'tanh'
                     }
        audio_args['fc_layers'] = [512, 64] if text else [512, 512]
    else:
        audio_args = None
    
    n_hiddens_list = [10] if text else [0]
    n_outputs = len(np.unique(Ttrain))
    
    nnet = tnn.TextNeuralNetworkClassifier(text_args, audio_args, n_hiddens_list, n_outputs, 
                                           h_activation_f='tanh', use_gpu=True, seed=1234)
    nnet.train(Xtrain if text else None, Atrain if audio else None, Ttrain, 
               validation_data=(Xval if text else None, Aval if audio else None, Tval),
               n_epochs=15, batch_size=32, learning_rate=0.001, opt='Adam'.lower(),
               shuffle=True, verbose=False)
    print(nnet)
    Y = nnet.use(Xtest if text else None, Atest if audio else None)
    return metrics.evaluate(Y, Ttest, verbose=False) # confmat, precision, recall, f1, accuracy
```

```python
train_args = [{'text': True, 'audio': False, 'text_mode': 'lstm', 'embed_mode': 'glove'},
              {'text': True, 'audio': False, 'text_mode': 'conv', 'embed_mode': 'glove'},
              {'text': False, 'audio': True, 'text_mode': None  , 'embed_mode': 'glove'},
              {'text': True, 'audio': True , 'text_mode': 'lstm', 'embed_mode': 'glove'}]
```

```python
task = 'emotion'

results = []
for i,args in enumerate(train_args):
    print(f'training {i+1}/{len(train_args)}')
    results.append(trainer(args['text'], args['audio'], args['text_mode'], args['embed_mode']))
```
